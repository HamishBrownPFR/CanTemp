import to pandas data frame

can specify multiple header rows, in this case skipping row 0 which is header

```python
import datetime
import pandas as pd
df=pd.read_csv('CR1000_AllData.dat',header=[1,2,3],parse_dates=True,index_col=0)
```

```python
df.shape
```

```python
df
```

for now, simplify the column names by just taking first row, discarding units and aggregation fields

```python
df.columns=[X[0] for X in df.columns]
df
```

Grab a data frame of the temperatures and plot these. **Note** label-based indexing.

```python
temps=df.ix[:,"SurfaceT_Plot_1":"SurfaceT_Plot_24"]
temps.plot(legend=False)
```

Have a look at the outlier plot...seems to be Plot 20 and inspection of raw data shows mostly no data for 20-24

```python
temp_mins=temps.describe().min()
temp_mins.sort()
temp_mins.head()
```

just look at plots 1:19 for now, plotting surface temp

Note use of comprehension for selecting columns with *SensBodyT*

```python
temps=df.loc[:,[X for X in temps.columns if 'SensBodyT' in X]]
temp_plot=temps.plot(legend=False,title='Surface Temp')
temp_plot.set_ylabel('Surface temp C')
```

```python
Calculate means in 11am-2pm 18 Feb
```

```python
temps['18-Feb 2012 11:00': '18-Feb 2012 14:00' ].mean()
```

Plot RH

```python
df.plot(y='RH_Max')
```

Just take a daytime slice

```python
df['18-Feb 2012 09:00': '18-Feb 2012 16:00' ].plot(y='RH_Max')
```

Take the same slice but plot 1h rolling average. **Note** need to remove NaN from series

```python
pd.rolling_median(df['18-Feb 2012 09:00': '18-Feb 2012 16:00' ]['RH_Max'],12).dropna().plot()
```

```python
pd.rolling_median(df['18-Feb 2012 09:00': '18-Feb 2012 16:00' ]['RH_Max'],12).dropna()['18-Feb 2012 10:00':'18-Feb 2012 11:00'].mean()
```
